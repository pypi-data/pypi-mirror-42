```python
%matplotlib inline
from lifelines import WeibullFitter, LogNormalFitter
import numpy as np
```

```python
T = np.random.exponential(10, 10000)
D = np.random.exponential(10, 10000)

keep = T > D
T = T[keep]
D = D[keep]
```

```python
(T > D).mean()
```

```python
keep.mean()
```

```python
from lifelines import NelsonAalenFitter
```

```python
wf = WeibullFitter().fit(T, entry=D)
naf = NelsonAalenFitter().fit(T, entry=D)
```

```python
wf.print_summary(3)
```

```python
ax=wf.plot(loc=slice(0, 10))
naf.plot(ax=ax, loc=slice(0, 10))
```
